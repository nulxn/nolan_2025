```java
// Popcorn Hack #1: Transform beginner code using compound assignment operators

public class GameStats {
    public static void main(String[] args) {
        int playerScore = 1000;
        int playerHealth = 100;
        int enemiesDefeated = 0;
        
        System.out.println("=== Initial Stats ===");
        System.out.println("Score: " + playerScore);
        System.out.println("Health: " + playerHealth);
        System.out.println("Enemies Defeated: " + enemiesDefeated);
        
        // Player defeats an enemy worth 250 points
        playerScore += 250;
        
        // Player takes 15 damage
        playerHealth -= 15;
        
        // Enemy count goes up
        enemiesDefeated += 1;
        
        System.out.println("\n=== After defeating enemy ===");
        System.out.println("Score: " + playerScore);
        System.out.println("Health: " + playerHealth);
        System.out.println("Enemies Defeated: " + enemiesDefeated);
        
        // Boss battle: double the current score!
        playerScore *= 2;
        
        System.out.println("\n=== After boss battle ===");
        System.out.println("Score: " + playerScore);
        
        // Healing potion restores health to 80% of current
        playerHealth *= 4;
        playerHealth /= 5;  // 4/5 = 0.8, but we need integer math
        
        System.out.println("\n=== After healing potion ===");
        System.out.println("Health: " + playerHealth);
        
        System.out.println("\n=== Final Stats ===");
        System.out.println("Score: " + playerScore);
        System.out.println("Health: " + playerHealth);
        System.out.println("Enemies Defeated: " + enemiesDefeated);
    }
}

GameStats.main(null);
```

```java
// Popcorn Hack #2: Program using at least 3 different compound assignment operators

public class ScoreTracker {
    public static void main(String[] args) {
        int score = 100;
        
        System.out.println("Starting score: " + score);
        
        // Deduct points for a wrong answer
        score -= 25;
        System.out.println("After wrong answer (-25): " + score);
        
        // Add bonus points for completing a challenge
        score += 50;
        System.out.println("After completing challenge (+50): " + score);
        
        // Double the score with a power-up
        score *= 2;
        System.out.println("After power-up (x2): " + score);
        
        // Divide score by difficulty level
        score /= 4;
        System.out.println("After difficulty adjustment (/4): " + score);
        
        // Find remainder after dividing by 7
        score %= 7;
        System.out.println("After modulo 7 operation (%7): " + score);
        
        System.out.println("\nFinal score: " + score);
    }
}

ScoreTracker.main(null);
```
